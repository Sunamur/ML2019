**Задача 1 (0.5 балла)**: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.

Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ – 3.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def nonzero_prod(X):
    a= np.diag(X)
    a.flags.writeable = True
    a[a==0]=1
    b =  np.prod(a)
    return b
```

```python
### test your code
X = np.arange(20).reshape(4, 5)
X[2, 2] = 0
assert nonzero_prod(X) == 6 * 18
print("OK")
```

**Задача 2 (0.5 балла)**: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит нулевой.

Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ – 5.

```python
def max_after_zero(x):
    a = np.where(x==0)[0]+1
    if len(x) in a:
        a = a[:-1]
    return x[a].max()
```

```python
### test your code
x = np.array([0, 1, 2, 0, 6, 0, 10, 0])
assert max_after_zero(x) == 10
print("OK")
```

**Задача 4 (1 балл)**: Даны две выборки объектов - X и Y (с одинаковым числом столбцов, объекты расположены по строкам). Вычислить матрицу евклидовых расстояний между объектами.

Для этой задачи реализуйте также функцию без использования numpy (с циклами). Сравните по скорости работы векторизованную версию с невекторизованной и с функцией scipy.spatial.distance.cdist на случайно сгенерированных матрицах размера 100 x 1000 (сравнения приведите ниже в ноутбуке).

```python
def cdist_numpy(X, Y):
    #Расширяем и вертим массивы для совместимости по размерам
    a = X.reshape(*X.shape,1)
    b=np.expand_dims(Y.T,2).reshape(1,X.shape[1],-1)
    # Разница-степень-сумма-корень
    res = np.sqrt(((a-b)**2).sum(axis=1))
    return res
```

```python
def cdist_python(X, Y):
    res = np.zeros((X.shape[0],Y.shape[0]))
    for x in range(X.shape[0]):
        for y in range(Y.shape[0]):
            res[x,y]=np.sqrt(((X[x]-Y[y])**2).sum())
    return res
```

```python
### test your code
from scipy.spatial.distance import cdist
X = np.arange(12).reshape(4, 3)
Y = np.arange(1, 16).reshape(5, 3)
assert np.allclose(cdist(X, Y), cdist_numpy(X, Y))
assert np.allclose(cdist(X, Y), cdist_python(X, Y))
print("OK")
```

```python
X=np.random.random((100,1000))
Y=np.random.random((100,1000))
```

```python
%%timeit
cdist_python(X,Y)
```

```python
%%timeit
cdist_numpy(X,Y)
```

```python
%%timeit
cdist(X,Y)
```

#### 11. [0.25 баллов] В каком количестве заказов есть товар, который стоит более 40% от суммы всего чека?

Возможно, будет удобно посчитать отдельно средний чек, добавить его в исходные данные и сделать необходимые проверки.

```python
df1 = pd.merge(#Создаем колонку с суммой чека для данного заказа
    df,
    df.groupby('order_id')['price_in_tab'].sum().reset_index().rename(columns={'price_in_tab':'order_sum'}),
    left_on='order_id',
    right_on='order_id',
    how='left'
)
df1['fraction_of_total'] = df1['price_atom']/df1['order_sum'] # Считаем долю каждого товара в заказе. Заметим, что мы здесь учитываем цену одного товара, а не всех одинаковых товаров.
print('Заказов с товаром, стоящим более чем 40% от всего заказа:')
#Отбираем все строки с долей от целого больше 40%, убираем дубликаты заказов и смотрим на размер таблицы
print(df1[df1['fraction_of_total']>0.4].drop_duplicates(subset=['order_id']).shape[0])
```

У matplotlib, конечно, же есть [документация](https://matplotlib.org/users/index.html) с большим количеством [примеров](https://matplotlib.org/examples/), но для начала достаточно знать про несколько основных типов графиков:
- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторое величины;
- scatter — график, показывающий взаимосвязь двух величин;
- bar — столбцовый график, показывающий взаимосвязь количественной величины от категориальной.

В этом задании вы попробуете построить каждый из них. Не менее важно усвоить базовые принципы визуализаций:
- на графиках должны быть подписаны оси;
- у визуализации должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах).
- сетка на графике помогает оценить значения в точках на глаз, это обычно полезно, поэтому лучше ее отрисовывать.

Работать мы будем с той же выборкой покупкок. Добавим новую колонку с датой покупки.

```python
import datetime

start = datetime.datetime(2018, 1, 1)
end = datetime.datetime(2018, 1, 31)
delta_seconds = int((end - start).total_seconds())

dates = pd.DataFrame(index=df.order_id.unique())
dates['date'] = [
    (start + datetime.timedelta(seconds=random.randint(0, delta_seconds))).strftime('%Y-%m-%d')
    for _ in range(df.order_id.nunique())]

# если DataFrame с покупками из прошлого заказа называется не df, замените на ваше название ниже
df1['date'] = df1.order_id.map(dates['date'])
```

#### 1. [0.5 баллов] Постройте гистограмму распределения сумм покупок и гистограмму средних цен отдельных видов продуктов item_name. 

Изображайте на двух соседних графиках. Для этого может быть полезен subplot.

```python
plt.rcParams['figure.figsize']=15,10
ax1 = plt.subplot(1,2,1)
df1.drop_duplicates(subset='order_id')['order_sum'].hist(bins=30, ax=ax1)
plt.title('Гистограмма сумм покупок')
plt.xlabel('Сумма чека')
plt.ylabel('Количество заказов')


ax2 = plt.subplot(1,2,2)
df1.groupby('item_name')['price_atom'].mean().hist(bins=30, ax=ax2)
plt.title('Гистограмма средних цен на отдельный товар')
plt.xlabel('Средняя стоимость товара')
plt.ylabel('Количество уникальных товаров')

plt.show()
```

#### 2. [0.5 баллов] Постройте график зависимости суммы покупок от дней.

```python
df1['date']=pd.to_datetime(df1['date'], format='%Y-%m-%d')
df1.groupby('date')['price_in_tab'].sum().plot()#.plot(x='date', y='order_sum')
plt.title('График зависимости суммы покупок от дня продажи')
plt.xlabel('Дата')
plt.ylabel('Выручка за день')
```

#### 3. [0,5 баллов] Постройте средних сумм покупок по дням недели (bar plot).

```python
df1['weekday'] = df1['date'].apply(lambda x: x.dayofweek)
weekday_dict = {0:'Понедельник',
1:'Вторник',
2:'Среда',
3:'Четверг',
4:'Пятница',
5:'Суббота',
6:'Воскресенье'}
df1['weekday'] = df1['weekday'].apply(lambda x: str(x)+'. '+weekday_dict[x])
df1.drop_duplicates(subset='order_id').groupby('weekday')['order_sum'].mean().plot(kind='bar')
plt.title('Среднняя выручка по дням недели')
plt.xlabel('День недели')
plt.ylabel('Средний чек, $')
```

#### 4. [0.5 баллов] Постройте график зависимости денег за товар от купленного количества (scatter plot).

```python
df1.plot(x='quantity', y='price_in_tab', kind='scatter')
plt.title('Зависимость общей цены от количества товаров')
plt.xlabel('Количество товаров')
plt.ylabel('Сумма за товар')
```
